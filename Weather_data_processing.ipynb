```python
df_weather=pd.DataFrame()
```

```python
predictor_columns=['windspeedKmph','windspeedMiles','WindGustKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover',
                  'DewPointF','tempF','WindChillF','humidity','time', 'date','airport']
```

```python
prev_filepath=filepath

for x in os.listdir():
    
    if x == "flight":
        continue
        
    
    filepath=prev_filepath
    filepath=filepath+"\\"+x
    os.chdir(filepath)
    for y in os.listdir():
        
        next_path=filepath+'\\'+y
        os.chdir(next_path)
        
        for z in os.listdir():
            a=os.getcwd()
            airport=a.split('\\')[-1]
            df=pd.read_json(z)
            x=df["data"]["weather"]
            y=pd.json_normalize(x,record_path="hourly",meta="date")
            y['airport']=airport
            y=y[predictor_columns]
            df_weather=pd.concat([df_weather,y])
            print(airport)
            
      
            
print("Done")
```
